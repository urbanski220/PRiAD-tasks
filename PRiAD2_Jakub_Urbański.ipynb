> **Zadanie 2** Wyświetl wykres słupkowy pokazujący liczbę ludności wszystkich dużych państw świata tj. liczących więcej niż 35 milionów mieszkańców w kolejności od najmniejszego do największego. 

```python
# rozwiązanie zadania
panstwa = pd.read_excel('alphabetical-list-of-countries2020.xlsx',header = 4,usecols = range(2,5), index_col = 0)
print(panstwa.head())
panstwa = panstwa.drop(columns=['Land Area'])
panstwa.columns = ['Ludnosc']
panstwa.index.name = 'Kraj'
print("\n", panstwa.head())
panstwa = panstwa[panstwa.Ludnosc > 35000000]
print("\n", panstwa.head())
panstwa = panstwa.sort_values(by='Ludnosc')
print("\n", panstwa.head())
panstwa = panstwa.T
plt.figure();
panstwa.iloc[0].plot(kind='bar');
```

> **Zadanie 3** Narysuj histogramy poszczególnych atrybutów zbioru danych `iris` z podziałem na klasy, oraz macierz wykresów punktowych. Oceń przydatność poszczególnych atrybutów do rozróżniania odmian irysa

```python
# rozwiązanie zadania
i = pd.read_csv('iris.csv', usecols = range(1,6))
sns.pairplot(i, kind="scatter", hue = "species")
plt.show()
```

**Wnioski:**

Po analizie histogramów można zauważyć, że atrybuty te mają stosunkowo mało wspólnych punktów dla poszczególnych gatunków irysa co ułatwia ich rozróżnienie dla konkretnych wartości.
